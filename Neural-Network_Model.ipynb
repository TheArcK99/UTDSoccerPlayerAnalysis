```python
X = df.drop('fee_cleaned', axis=1)
y = df['fee_cleaned']
```

K-fold Cross Validation

```python
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
k_folds = 10
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

param_grid = {
    'hidden_layer_sizes': [(70, 70), (80, 80), (90, 90)],
    'activation': ['tanh'],
    'max_iter': [x for x in range(0, 1501, 100)],
}

mse_scores = []
r2_scores = []
```

```python
neural_network_model = MLPRegressor(activation='tanh', hidden_layer_sizes=(100, 100, 100, 100), max_iter=1000, random_state=42)

for train_index, test_index in kf.split(X):
    features_train, features_test = X.iloc[train_index], X.iloc[test_index]
    target_train, target_test = y.iloc[train_index], y.iloc[test_index]

    neural_network_model.fit(features_train, target_train)

    target_pred = neural_network_model.predict(features_test)

    mse = mean_squared_error(target_test, target_pred)
    r2 = r2_score(target_test, target_pred)

    mse_scores.append(mse)
    r2_scores.append(r2)

average_mse = np.mean(mse_scores)
average_r2 = np.mean(r2_scores)

print("Average Mean Squared Error:", average_mse)
print("Average R-squared:", average_r2)
```
