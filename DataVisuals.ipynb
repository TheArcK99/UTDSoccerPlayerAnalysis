Importing Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

Using pandas to convert csv file into data frame

```python
df = pd.read_csv('/content/TotalFBREFPlayerData - Sheet1.csv')
df = df.dropna()
display(df)
```

**ATTACKERS**

Filter the attacking positions

```python
attackingDF = df[(df['position'] == 'Centre-Forward')
| (df['position'] == 'Second Striker')
| (df['position'] == 'Left Winger')
| (df['position'] == 'Right Winger')]
display(attackingDF)
```

Filter out unwanted columns

```python
dropLst = ['club_involved_name', 'club_name', 'transfer_movement', 'season', 'pass_complete%', 'tackle', 'block', 'goals_against', 'goals_against90', 'Saves', 'Saves%', 'PSxG', 'nationality']
for i in range(len(dropLst)):
  attackingDF = attackingDF.drop(dropLst[i], axis='columns')
display(attackingDF)
```

Goals vs xG

```python
sns.regplot(x = 'assists', y = 'fee_cleaned', data = attackingDF)
```

Goals vs Transfer Fee

```python
sns.regplot(x = 'goals', y = 'fee_cleaned', data = attackingDF)
```

Age vs Transfer Fee

```python
sns.regplot(x = 'age', y = 'fee_cleaned', data = attackingDF)
```
